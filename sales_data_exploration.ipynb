```python
df.dropna(axis=1) #remove the column that contain null value
```

```python
df["Sales_Amount"].filna(df["Sales_Amount"].mean()) #fill the column null value with mean
```

```python
df["Sales_Amount"].filna(df["Sales_Amount"].mean()) #fill the column null value with median
```

```python
df["Sales_Amount"].fillna(df["Sales_Amount"].mode()[0]) #fill the column null value with mode
```

```python
df.duplicated()
```

```python
df.drop_duplicated() #to drop the duplicat row
```

```python
df.dtypes #check the datatype of the columns in dataset
```

```python
df["Sales_Amount"] = df["Sales_Amount"].astype(int) #to change the datatype of column
```

```python
df.rename(columns={"Sales_Amount":"Amount"}) #to change or rename the column name
```

```python
df.drop(axis=1) #to drop the unnecessary column from the dataset
```

```python
df["Sales_Rep"] = df["Sales_Rep"].str.lower() #to make string in lowercase
```

```python
df["Sales_Rep"] = df["Sales_Rep"].str.strip() #to remove space in the string if available
```

```python
df["Sales_Amount"] = df[df["Sales_Amount"] > 0] #filtering the data amount can not be negative
```
